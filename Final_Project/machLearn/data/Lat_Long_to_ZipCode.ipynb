```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('baltimore.csv')
data.head()
```

```python
data['zipcode']=""
data.head()
```

```python
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)
```

```python
for index, row in data.iterrows():
    lat = row['Latitude']
    long = row['Longitude']
    zipcode = search.by_coordinates(lat, long, returns = 1) 
    if len(zipcode) > 0:
        blah = zipcode[0].zipcode
    else:
        blah = 0
    data.at[index, 'zipcode'] = blah
    if index%100 == 0:
        print("still going")
```

```python
table2 = data['zipcode'].value_counts().rename_axis('zipcode').reset_index(name='total_crimes')
table2.head()
```
