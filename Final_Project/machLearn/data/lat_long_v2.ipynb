```python
new_table['zipcode'] = ""
new_table.head()
```

```python
new_table['Latitude'] = pd.to_numeric(new_table['Latitude'])
new_table['Longitude'] = pd.to_numeric(new_table['Longitude'])

new_table.Longitude.dtype
```

```python
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)
```

```python
for index, row in new_table.iterrows():
    lat = row['Latitude']
    long = row['Longitude']
    zipcode = search.by_coordinates(lat, long, returns = 1) 
    if len(zipcode) > 0:
        blah = zipcode[0].zipcode
    else:
        blah = 0
    new_table.at[index, 'zipcode'] = blah
    if index%100 == 0:
        print(str(index) + " still going")
```

```python
for index, row in table2.iterrows():
    look = row['zipcode']
    total = 0
    for index2, row2 in new_table.iterrows():
        if row2.zipcode == look:
            total = total + 1
    table2.at[index, 'total_crimes'] = total
#     print(row['total_crimes'])
```
