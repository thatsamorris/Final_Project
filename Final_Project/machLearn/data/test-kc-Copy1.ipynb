```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('raw city data/detroit.csv')
data.head()
# data.shape
```

```python
# data['location_block'].value_counts()

# shows how many crimes occurred in each zipcode
new_table = data['Location'].value_counts().rename_axis('Location').reset_index(name='total_crimes')
# new_table['Location']=new_table['Location'].str.split(',',expand=True)[1]
# new_table.head()
new_table['lat']=new_table['Location'].str.split(',',expand=True)[0]
new_table['long']=new_table['Location'].str.split(',',expand=True)[1]
new_table['lat']=new_table['lat'].str.split('(',expand=True)[1]
new_table['long']=new_table['long'].str.split(')',expand=True)[0]

new_table['lat']=pd.to_numeric(new_table['lat'])
new_table['long']=pd.to_numeric(new_table['long'])

new_table.head()
```

```python
data.shape
```

```python
new_table.shape
```

```python
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)

for index, row in new_table.iterrows():
    lat = row['lat']
    long = row['long']
    zipcode = search.by_coordinates(lat, long, returns = 1) 
    if len(zipcode) > 0:
        blah = zipcode[0].zipcode
    else:
        blah = 0
    new_table.at[index, 'zipcode'] = blah
    if index%200 == 0:
        print(str(index) + " still going")
```

```python
# new_table.head()
table2 = new_table.drop(['Location','lat','long'], axis=1)
d = {'total_crimes': 'sum'}
tb2_new = table2.groupby(table2['zipcode']).aggregate(d).reset_index()
tb2_new = tb2_new.sort_values(by=['total_crimes'], ascending=False)
tb2_new.head()
```

```python
tb2_new.zipcode = pd.to_numeric(tb2_new.zipcode)
tb2_new.zipcode.dtype
```

```python
census = pd.read_csv('census_data.csv')
census.head()
# new_table.head()

# new_table.zipcode.dtype
```

```python
tb2 = tb2_new.merge(census, left_on='zipcode', right_on='Zipcode', how='left')
tb2.head()
```

```python
tb2['crime rate'] = tb2.total_crimes / tb2.Population
tb2.sort_values(by=['crime rate'], ascending=False)
tb2 = tb2.reset_index(drop = True)
tb2.head()
```

```python
tb2 = tb2.drop(['Zipcode'], axis=1)
tb2.head()
```

```python
tb2 = tb2.drop([0,1,53])
tb2 = tb2.sort_values(by=['crime rate'], ascending=False).reset_index(drop=True)
tb2.head(500)
```
