```python
census = census.sort_values(by=['crime_rate'])
census.head()
```

```python
census['crime_rating'] = ""
census = census.reset_index(drop=True)
census.head()
```

```python
max_num = census['crime_rate'].max()
min_num = census['crime_rate'].min()
print(max_num)
print(min_num)
```

```python
# census['encode'] = ""
# census.head()
for index, row in census.iterrows():
    if(index < 142):
        blah = "low"
    elif(index < 284):
        blah = "medium"
    else:
        blah = "high"
    census.at[index, 'crime_rating'] = blah
census.head()
```

```python
census['encode'] = ""
# census.head()
for index, row in census.iterrows():
    if(row['crime_rating'] == "low"):
        blah = 0
    elif(row['crime_rating'] == "medium"):
        blah = 1
    else:
        blah = 2
    census.at[index, 'encode'] = blah
census.head()
```

```python
census.to_csv('combined_cities_census.csv', encoding='utf-8', index=False)
```

```python
X = census[["Median Age", "Household Income", 
            "Per Capita Income", "Poverty Rate"]]
y = census["encode"].values.reshape(-1, 1)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=4)
```

```python
 from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

```python
from keras.utils import to_categorical

# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
print(y_train_categorical.shape)
print(y_test_categorical.shape)
```

```python
from keras.models import Sequential

model = Sequential()
```

```python
from keras.layers import Dense
number_inputs = 4
number_hidden_nodes = 8
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
number_classes = 3
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
 # Use categorical crossentropy for categorical data and mean squared error for regression
# Hint: your output layer in this example is using software for logistic regression (categorical)
# If your output layer activation was `linear` then you may want to use `mse` for loss
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=500,
    shuffle=True,
    verbose=5
)
```

```python
 # Evaluate the model using the testing data
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=5)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# predicting outcome of data
import numpy as np
new_data = np.array([[0.2, 0.3, 0.4, 0.5]])
print(f"Predicted class: {model.predict_classes(new_data)}")
```
