```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
census = pd.read_csv('data/census_data.csv')
census.head()
```

```python
crime = pd.read_csv('data/KCPD_Crime_Data_2017.csv')
crime.head()
crime.shape
```

```python
crime.Location.head()
```

```python
crime[['Address 1', 'Address 2', 'Coordinates']] = crime['Location'].str.split('\n',expand=True)
```

```python
crime.head()
```

```python
crime['Zipcode'] = ""
crime.head()
```

```python
test=crime['Address 2'].str.split('KANSAS CITY',expand=True)[1]
crime['Zipcode']=test.str.split(' ',expand=True)[1]
```

```python
crime.loc[77313,'Zipcode'] = 64114
crime['Zipcode'] = pd.to_numeric(crime['Zipcode'])
```

```python
main = crime.merge(census, left_on='Zipcode', right_on='Zipcode', how='left')
main.head()
```

```python
main.shape
# 132139
```

```python
# main.to_csv('combined.csv')
```

```python
# shows how many crimes occurred in each zipcode
new_table = main['Zipcode'].value_counts().rename_axis('Zipcode').reset_index(name='total_crime')
new_table.head()
# new_table.shape
```

```python
new_table = new_table.merge(census, left_on='Zipcode', right_on='Zipcode', how='left')
# new_table.shape
new_table.head()
```

```python
#dropped rows with no data
test = new_table.dropna()
test.shape
```

```python
test["crime_rate"] = ""
test["crime_rate"] = test["total_crime"]/test["Population"]
test.head()
```

```python
array = ["Population", "Median Age", "Household Income", "Per Capita Income", "Poverty Count", "Poverty Rate"]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

for variable in array:
    X = test[[variable]]
    y = test["total_crime"].values.reshape(-1, 1)

    # Fitting our model with all of our features in X
    model.fit(X, y)

    score = model.score(X, y)
    print(f"{variable} - R2 Score: {score}")
```

```python
X = test[["Population", "Median Age", "Household Income", "Per Capita Income", "Poverty Count", "Poverty Rate"]]
y = test["total_crime"].values.reshape(-1, 1)

# Fitting our model with all of our features in X
model.fit(X, y)

score = model.score(X, y)
print(f"R2 Score (Median Age, Household Income, Per Capita Income, Poverty Count): {score}")
```
